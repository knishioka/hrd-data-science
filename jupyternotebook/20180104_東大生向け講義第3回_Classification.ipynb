```python
import pandas as pd
import seaborn as sns
from sklearn.svm import SVC
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
for l in iris.species.unique():
    target = iris[iris.species == l]
    plt.scatter(target.sepal_length, target.sepal_width, label=l)
plt.legend()
```

```python
# sliceでx(説明変数), y(目的変数)を取得
# x_all = iris[['sepal_length', ''sepal_width, 'petal_length', 'petal_width']]でも同様
x_all = iris.iloc[:, :-1]
y_all = iris.species
```

```python
svc = SVC()
svc.fit(x_all, y_all)
svc.predict(x_all)
```

```python
from sklearn.metrics import accuracy_score
```

```python
# trainingとtestを同じものとしている例。
accuracy_score(y_all, svc.predict(x_all))
```

```python
train_size = 100
x_train = iris.iloc[:train_size, :-1]
y_train = iris.iloc[:train_size, -1]
x_test = iris.iloc[train_size:, :-1]
y_test = iris.iloc[train_size:, -1]
```

```python
svc.fit(x_train, y_train)
accuracy_score(svc.predict(x_test), y_test)   # サンプルの順番の影響をうける。今回はデータが種類で順番に並んでいた
```

```python
shuffled = iris.sample(frac=1)
shuffled
```

```python
x_shuffled_tr = shuffled.iloc[:train_size, :-1]
y_shuffled_tr = shuffled.iloc[:train_size, -1]
x_shuffled_te = shuffled.iloc[train_size:, :-1]
y_shuffled_te = shuffled.iloc[train_size:, -1]
```

```python
svc.fit(x_shuffled_tr, y_shuffled_tr)
accuracy_score(y_shuffled_te, svc.predict(x_shuffled_te))
```

```python
# 先程は手動でshuffle, train dataとtest dataへの振り分けを行ったが、sklearnのtrain_test_splitで簡便に行える
from sklearn.model_selection import train_test_split
x_tt_tr, x_tt_te, y_tt_tr, y_tt_te = train_test_split(iris.iloc[:, :-1], iris.iloc[:, -1])
```

```python
svc.fit(x_tt_tr, y_tt_tr)
accuracy_score(svc.predict(x_tt_te), y_tt_te)
```

```python
# cross_val_scoreはaccuracyの評価まで一緒に行えて、さらに簡便
from sklearn.model_selection import cross_val_score
```

```python
result = cross_val_score(SVC(), iris.iloc[:, :-1], iris.iloc[:, -1], cv=5)
result
```

```python
result.mean(), result.std()
```

```python
from sklearn.naive_bayes import GaussianNB
cross_val_score(GaussianNB(), iris.iloc[:, :-1], iris.iloc[:, -1], cv=5)
```

```python
titanic = sns.load_dataset('titanic').dropna()
```

```python
titanic.head()
```

```python
# 問題なくfitできる
gnb = GaussianNB()
gnb.fit(titanic[['age', 'fare']], titanic['survived'])
```

```python
# sexやclassに文字が入っているので、fitの際にエラーになる
gnb = GaussianNB()
gnb.fit(titanic[['age', 'fare', 'sex']], titanic['survived'])
```

```python
pd.get_dummies(titanic.sex).head(3)
```

```python
df1 = pd.concat([titanic[['age', 'fare']],
                pd.get_dummies(titanic.sex)],
               axis=1)
```

```python
df1.head()
```

```python
gnb1 = GaussianNB()
gnb1.fit(df1[['age', 'fare', 'female', 'male']], titanic['survived']) # dummy変数を使うとfitできる
```

```python
# sexを説明変数に入れる前と後で予測精度を比較
```

```python
accuracy_score(titanic.survived, gnb.predict(titanic[['age', 'fare']]) )
```

```python
accuracy_score(titanic.survived, gnb1.predict(df1[['age', 'fare', 'female', 'male']]) )
# 性別を入れることで予測精度が少し上がった
```
