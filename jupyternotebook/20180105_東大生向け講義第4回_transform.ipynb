```python
%matplotlib inline

import pandas as pd
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.plot.scatter('sepal_length', 'sepal_width')
```

```python
# StandardScaler：標準化
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss
```

```python
s_scaled = pd.DataFrame(ss.fit_transform(iris.iloc[:, :-1]))
s_scaled
```

```python
s_scaled.describe() # meanがほぼ0に、stdはほぼ1になっている
```

```python
# MinMaxScaler：正規化
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler()
mms
```

```python
m_scaled = pd.DataFrame(mms.fit_transform(iris.iloc[:, :-1]))
m_scaled
```

```python
m_scaled.describe() # minが0に、maxが1になっている
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
fig = plt.figure()
ax = Axes3D(fig)

ax.scatter(iris.sepal_length, iris.sepal_width, iris.petal_length)
```

```python
# PCA(Principal Component Analysis, 主成分分析)により、次元削減を行う
# 主成分を与える変換は、第一主成分の分散を最大化する。
# 続く主成分はそれまでに決定した主成分と直交するように選ばれることで、分散を最大化させていく
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
```

```python
pca.fit_transform(iris.iloc[:, :-1])
```

```python
reduced = pca.fit_transform(iris.iloc[:, :-1])
plt.scatter(reduced[:, 0], reduced[:, 1])
```

```python
# 次サイトよりdataをDownloadする必要あり→https://data.world/crowdflower/disasters-on-social-media
df = pd.read_csv('~/Downloads/socialmedia-disaster-tweets-DFE.csv', error_bad_lines=False)
df.head()
```

```python
# CountVectorizer でテキスト内の単語の出現回数をカウントできる
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
df._golden.value_counts()
```

```python
# fitとtransformを同時に行う
transformed = cv.fit_transform(df.text)
transformed
```
