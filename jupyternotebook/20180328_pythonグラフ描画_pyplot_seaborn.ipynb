```python
# Mercari datasetを使う。下記URLのtrain.tsv.7zをダウンロードする
# https://www.kaggle.com/c/mercari-price-suggestion-challenge/data
data = pd.read_csv('mercari_train.tsv.gz', delimiter='\t', index_col='train_id')
data['length_desc'] = data.item_description.apply(lambda x: len(str(x)))
data
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
x = data.item_condition_id.unique()
x.sort()
y = [data[data.item_condition_id == c].price.mean() for c in x]
ax.bar(left=x, height=y, color='orangered')
ax.set_xlabel('item_condition', size=18)
ax.set_ylabel('mean_price', size=18)
ax.legend()
```

```python
categories = data.category_name.value_counts(ascending=False).index
categories
```

```python
w_shoes_categries = []
for c in categories:
    if c.find('Women/Shoes/') > -1:
        w_shoes_categries.append(c)
w_shoes_categries
top9_w_shoes_categries = w_shoes_categries[:9]
top9_w_shoes_categries
```

```python
shoes_mean = data[data.category_name.isin(w_shoes_categries)].price.mean()
shoes_mean
```

```python
fig, axes = plt.subplots(3, 3, figsize=(24, 16))
fig.subplots_adjust(wspace=0.2, hspace=0.4)
for i, c in enumerate(top9_w_shoes_categries):
    df = data[data.category_name == c]
    x = df.item_condition_id.unique()
    x.sort()
    y = [df[df.item_condition_id == s].price.mean() for s in x]
    axes[i//3, i%3 - 1].bar(left=x, height=y)
    axes[i//3, i%3 - 1].set_title(f"'{c.split('/')[-1]}' price by condition", fontsize=20)
    axes[i//3, i%3 - 1].set_xlabel('item condition', fontsize=16)
    axes[i//3, i%3 - 1].set_ylabel('price', fontsize=16)
    axes[i//3, i%3 - 1].set_ylim(0, 75)
    axes[i//3, i%3 - 1].axhline(y=shoes_mean, color='red')
    axes[i//3, i%3 - 1].set_xticks([1, 2, 3, 4, 5])
    axes[i//3, i%3 - 1].set_xticklabels(['new', 'mint', 'nearly mint', 'good', 'acceptable'], rotation=30, size=14)
```

```python
data['middle_category'] = data.category_name.apply(lambda x: x.split('/')[0] + '/' + x.split('/')[1] if type(x) == str else None)
data
```

```python
top10_middle_categories = data.groupby('middle_category').size().sort_values(ascending=False).head(10).index
top10_middle_categories
```

```python
from itertools import product
from collections import defaultdict
mean_price_by_middle_category = data.groupby(['middle_category', 'item_condition_id']).price.mean()
dic = defaultdict(dict)
for c1, c2 in product([1, 2, 3, 4, 5], repeat=2):
    if c1 == c2:
        dic[c1][c2] = None
    else:
        dic[c1][c2] = mean_price_by_middle_category['Beauty/Makeup'][c1] - mean_price_by_middle_category['Beauty/Makeup'][c2]
input_data = pd.DataFrame.from_dict(dic, orient='index')
```

```python
# ヒートマップ①：matplotlib
# https://matplotlib.org/api/_as_gen/matplotlib.pyplot.pcolor.html
fig, ax = plt.subplots()
heatmap = ax.pcolor(input_data.sort_index(ascending=False), vmin=input_data.min().min(), vmax=input_data.max().max(), cmap="GnBu")
fig.colorbar(heatmap)
for x, y in product([1, 2, 3, 4, 5], repeat=2):
    if 6 - x == y:
        pass
    else:
        plt.text(x - 0.7, y - 0.5, -round(input_data.sort_index(ascending=False).loc[x, 6 - y], 2), color='red')
ax.set_xticks([0.5, 1.5, 2.5, 3.5, 4.5])
ax.set_xticklabels([1, 2, 3, 4, 5])
ax.set_yticks([0.5, 1.5, 2.5, 3.5, 4.5])
ax.set_yticklabels([5, 4, 3, 2, 1])
ax.set_title('Beauty/Makeupの商品のcondition同士の価格差')
# 指定できる色一覧。https://matplotlib.org/examples/color/colormaps_reference.html
```
